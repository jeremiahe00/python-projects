```python
gameboard = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']

def display(board):
    print(board[0] + ' | ' + board[1] + ' | ' + board[2])
    print('---------')
    print(board[3] + ' | ' + board[4] + ' | ' + board[5])
    print('---------')
    print(board[6] + ' | ' + board[7] + ' | ' + board[8])
```

```python
choices = ['X', 'O']
player1 = ''
player2 = ''

def choice_pick():
    
    choice = input('Would you like to be X or O: ')
    
    while choice not in choices:
        print('This is not a valid choice')
        
        choice = input('Would you like to be X or O: ')
        
        if (choice in choices):
            break
    
    return choice    
```

```python
def clear_board(gameboard):
    for num in range(len(gameboard)):
        gameboard[num] = ' '
    #gameboard = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
```

```python
positions = list(range(0,9))

def position_pick(player):
    print('This is for visual representation only')
    print('0|1|2')
    print('3|4|5')
    print('6|7|8')
    pos = int(input('Please pick a position from 0 - 8: '))
    
    while pos not in positions or gameboard[pos] != ' ':
        print('This is not a valid choice')
        
        pos = int(input('Please pick a position from 0 - 8: '))
        
        if (pos in positions and gameboard[pos] == ' '):
            break
    
    gameboard[pos] = player
```
